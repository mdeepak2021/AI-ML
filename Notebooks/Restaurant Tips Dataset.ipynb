### Restaurant Tips Dataset

This dataset is available by default within the Python Seaborn library and can also be downloaded from different datasources available over the internet

The 'Tips' dataset, embedded within Seaborn, is a collection of records representing the tipping behavior of restaurant patrons. The dataset comprises:

- `total_bill`: Total bill amount.
- `tip`: Tip amount.
- `sex`: Gender of the customer.
- `smoker`: Whether the customer is a smoker.
- `day`: Day of the week.
- `time`: Whether itâ€™s lunch or dinner.
- `size`: Number of people in the party.

```python
# Importing the Python libraries

import pandas as pd
import numpy as np
import seaborn as sns
```

Loading the Tips dataset from seaborn

```python
tips = sns.load_dataset("tips")
tips.head()
```

### Data Visualization

We will start visualising the dataset using different plots available in Seaborn and come up with our analysis

```python
# Scatterplot
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
# Relplot
sns.relplot(x="total_bill", y="tip", data=tips);
```

```python
# Relplot with hue
sns.relplot(x="total_bill", y="tip", hue="time", data=tips);
```

From the above graph, most of time, the total bill for dinner was greater than lunch, looks like people prefer to have dinner than lunch in this restuarant

```python
sns.relplot(x="total_bill", y="size", hue="smoker", style="smoker", data=tips);
```

```python
sns.relplot(x="total_bill", y="tip", hue="time", style="smoker", data=tips);
```

The people visitng to the restaurant where smoker than non-smokers visiting the restaurant

```python
sns.relplot(x="total_bill", y="tip", size="size", data=tips);
```

The higher the total bill, more generous tip was given

```python
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15, 200), data=tips);
```

```python
# Lineplot
sns.lineplot(x="total_bill", y="tip", data=tips)
```

```python
g = sns.relplot(x="total_bill", y="tip", kind="line", data=tips)
```

```python
palette = sns.cubehelix_palette(light=.8, n_colors=2)
sns.relplot(x="total_bill", y="tip",
            hue="sex", style="size",
            palette=palette,
            kind="line", data=tips);
```

From the above graph, Men seems to be spending more on the restaurant and also generous interms of giving tips

```python
df = pd.DataFrame(dict(time=pd.date_range("2017-1-1", periods=500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df)
g.fig.autofmt_xdate()
```

```python
sns.relplot(x="size", y="tip", hue="time", col="smoker", kind="line", data=tips);
```

```python
sns.relplot(x="total_bill", y="tip", hue="time", 
            col="day", col_wrap=5,
            height=3, aspect=.75, linewidth=2.5,
            kind="line", data=tips.query("sex == 'Male'"));
```

From above graph, seems like weekends (saturday & sunday) is when people prefer to go to the restaurant and that too for dinner
